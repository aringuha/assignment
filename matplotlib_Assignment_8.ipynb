**Scipy: We have the min and max temperatures in a city In India for each months of the year.
We would like to find a function to describe this and show it graphically.**

```python
import matplotlib.pyplot as plt
from scipy import optimize
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
max_temp = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
min_temp = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])
months = np.arange(12)
```

**1. fitting it to the periodic function**

```python
def yearly_temps(times, avg, ampl, time_offset):
    return (avg + ampl * np.cos((times + time_offset) * 2 * np.pi / times.max()))
```

```python
res_max, cov_max = optimize.curve_fit(yearly_temps, months, max_temp, [20, 10, 0])
res_min, cov_min = optimize.curve_fit(yearly_temps, months, min_temp, [-40, 20, 0])
```

**2. plot the fit**

```python
days = np.linspace(0, 12, num=365)

plt.figure(figsize=(8,6))
plt.plot(months, max_temp, 'ro')
plt.plot(days, yearly_temps(days, *res_max), 'r-')
plt.plot(months, min_temp, 'bo')
plt.plot(days, yearly_temps(days, *res_min), 'b-')
plt.xlabel('Month')
plt.ylabel('Temperature ($^\circ$C)')

plt.show()
```

**1. Matplotlib: Create a pie chart presenting the male/female proportion**

```python
url = 'https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
titanic = pd.read_csv(url)
titanic.head()
```

```python
gender = titanic['sex'].astype('category').cat.codes #convert to series of numerical values
plt.pie(gender.value_counts(), labels = titanic['sex'].unique(), autopct='%1.1f%%')
plt.title('Male/Female Proportion')
plt.axis('equal')
plt.show()
```

**2. Matplotlib: Create a scatterplot with the Fare paid and the Age, differ the plot color by gender**

```python
plt.figure(figsize=(8,6))
plt.scatter(titanic['fare'], titanic['age'], c = gender, cmap='RdBu')
plt.xlabel("Fare")
plt.ylabel("Age")
plt.show()
```
